# `Querying an Array`
Match an array field :-

{ 

....

\<field\> : [\<value\>, \<value\>, ...]

...

}

----

[Array query operators](https://docs.mongodb.com/manual/reference/operator/query-array/) :- For querying on an array field.

- [$all](https://docs.mongodb.com/manual/reference/operator/query/all/#mongodb-query-op.-all) - Matches arrays that contain all elements specified in the query.

- [$size](https://docs.mongodb.com/manual/reference/operator/query/size/#mongodb-query-op.-size) - Selects documents if array field is a specified size

[Array projection  operators](https://docs.mongodb.com/manual/reference/operator/projection/_) :- Project elements from an array field

- [$slice](https://docs.mongodb.com/manual/reference/operator/projection/slice/#proj._S_slice) - Limits number of elements projected from an array.

## Querying using array operators

----

Query an array that contains given value

For example, 
* Find documents where `products` array contains `Derivatives` element


```python
# Find documents that contains specific array element

cur = db.accounts.find(
                        # query expression
                        {
                            'products':'Derivatives'
                        },
                        # projection
                        {
                            'products':1,
                            '_id':0
                        })

# Print documents
for doc in cur:
    pp.pprint(doc)
```

---

Query a collection that contains a documents **with specific array values only and in mentioned order**.

For example, 
* Find documents that contains only elements `[Derivatives, InvestmentStock]` and in same order

```python
# Document that contains exact array values - same values and same order

cur = db.accounts.find(
                        # query expression
                        {
                            'products':['Derivatives', 'InvestmentStock']
                        },
                        # projection
                        {
                            'products':1,
                            '_id': 0
                        })

# Print documents
for doc in cur:
    pp.pprint(doc)
```


**`$all` array operator**

[$all](https://docs.mongodb.com/manual/reference/operator/query/all/#op._S_all) matches arrays that contain all elements specified in the query.

**Syntax -** `{ <field>: { $all: [ <value1> , <value2> ... ] } }`

----

For example, 
* Find documents that contain `Commodity` and `InvestmentStock` in `products` array field
    * Documents with array fiel containing at least these elements will be retrieved
    * Order does not matter here

```python
# Documents that contains multiple array elements in any order

cur = db.accounts.find(
                        # query expression
                        {
                            'products':{
                                        '$all':['InvestmentStock', 'Commodity']
                                        }
                        },
                        # projection
                        {
                            'products':1,
                            '_id':0
                        })

# Print documents

for doc in cur:
    pp.pprint(doc)
```

**`$size` array operator**

[$size](https://docs.mongodb.com/manual/reference/operator/query/size/#op._S_size) selects documents if the array field is a specified size.

**Syntax -** `{ <field>: { $size: <size> } }`

---

For example, 
* Retrieve all documents that have only two elements in `products` array field

```python
# Documents with array of size 2

cur = db.accounts.find(
                        # query expression
                        {
                            'products':{
                                        '$size':2
                                        }
                        },
                        # projection
                        {
                            'products':1,
                            '_id':0
                        })

# Print documents

for doc in cur:
    pp.pprint(doc)
```

**`$slice` projection operator**

Using [$slice](https://docs.mongodb.com/manual/reference/operator/projection/slice/#proj._S_slice) you can limit elements of an array to return in query result

**Syntax -** `find(<query>,{ <arrayField>: { $slice: <number> } })`

----

For instance, 
* Return only first two elements of `products` array where size of array is 5

```python
# Limit the number of array elements to return

cur = db.accounts.find(
                        # query expression
                        {'products':{
                                        '$size':5
                                    }
                        },
                        # projection
                        {
                            'products':{'$slice':2},
                            '_id':0
                        })

# Print documents

for doc in cur:
    pp.pprint(doc)
```


* **To return the last element(s) of array, prefix the minus(-) operator before the number**


```python
# Complete document
pp.pprint(
    db.accounts.find_one({
                            'products':{
                                            '$size':5
                                        }
                        })
)
```

```python
# Return last 2 array element

pp.pprint(
    db.accounts.find_one(
                        # query expression
                        {
                            'products':{
                                            '$size':5
                                        }
                        },
                        # projection
                        {
                            'products':{
                                            '$slice':-2
                                        }
                        })
)
```

**Also returns array elements after skipping specified number of elements**

For example, 
* Retrieve 3rd element of `products` field where `account_id` is 599752

```python
# Original doc
pp.pprint(
    db.accounts.find_one({
                            'account_id': 599752
                        })
)
```

```python
# Return third array element
# skip 2 and return 1 element

pp.pprint(
    db.accounts.find_one(
                            # query expression
                            {
                                'account_id': 599752
                            },
                            # projection
                            {
                                'products':{
                                                '$slice': [2, 1]
                                            }
                            })
)
```

---
**Query array based on index of element.**

For example, 
* Retrieve those documents where first element of `products` array is `Derivatives`

```python
# First element `Derivatives`
cur = db.accounts.find(
                        # query expression
                        {
                            'products.0': 'Derivatives', 
                        },
                        # projection
                        {
                            'products': 1,
                            '_id':0
                        })

for doc in cur:
    pp.pprint(doc)
```

For example, 
* Retrieve those documents where first element of `products` array is `Derivatives`

```python
# Second element `Derivatives`
cur = db.accounts.find(
                        # query expression
                        {
                            'products.1': 'Derivatives', 
                        },
                        # projection
                        {
                            'products': 1,
                            '_id': 0
                        })

for doc in cur:
    pp.pprint(doc)
```
